# Necessary imports are made

```python
import time
import spotipy
from spotipy.oauth2 import SpotifyOAuth
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
#for graphs
import seaborn as sns
import matplotlib.pyplot as plt
import json
```

# Creating the data frame
The data frame is created from, Streaming_History_Audio_2018-2024, that I downloaded from spotify.

```python
# Load Spotify data
with open('Streaming_History_Audio_2018-2024.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
    
df = pd.DataFrame(data)

df['ts'] = pd.to_datetime(df['ts'])

df['Year'] = df['ts'].dt.year
#for the time zone error
df['ts'] = df['ts'].dt.tz_localize(None)   
df['Month'] = df['ts'].dt.to_period('M')

yearly_listening = df.groupby('Year')['ms_played'].sum() / (1000 * 60 * 60)  # Convert ms to hours

plt.figure(figsize=(10, 6))
ax = yearly_listening.plot(kind='bar', color='orange')
plt.title('Total Listening Time by Year')
plt.ylabel('Total Listening Time (hours)')
plt.xlabel('Year')
plt.xticks(rotation=45)
plt.tight_layout()

# Add exact values above the bars
for index, value in enumerate(yearly_listening):
    plt.text(index, value + 0.5, f"{value:.1f} hrs", ha='center', va='bottom', fontsize=10)

plt.show()
```

# Number of skipped songs per month
I constructed a histogram.

```python
skipped_songs = df[df['skipped'] == True]

# Count the number of skipped songs per month
skipped_counts = skipped_songs.groupby('Month').size()

ax = skipped_counts.plot(kind='bar', figsize=(12, 6), color='orange')
plt.title('Number of Skipped Songs Per Month')
plt.ylabel('Number of Skipped Songs')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.tight_layout()

# Annotate each bar with the exact number of skipped songs
for i, count in enumerate(skipped_counts):
    plt.text(i, count + 1, str(count), ha='center', va='bottom', fontsize=10)
    
plt.show()
```

# Analysis of Skipped Songs Based on Stress Levels

**Objective**:  
The goal of this analysis is to examine whether there is a significant difference in the number of skipped songs during stressful and non-stressful periods. Stressful months are predefined, and the analysis focuses on the mean number of skipped songs in these two scenarios.

**Null Hypothesis (H)**:  
There is no significant difference in the average number of skipped songs between stressful and non-stressful periods.

**Alternative Hypothesis (H)**:  
There is a significant difference in the average number of skipped songs between stressful and non-stressful periods, implying that stress levels might influence skipping behavior.

**Interpretation**:  
The p-value (0.164) is greater than the commonly used significance level (e.g., 0.05). This indicates that we fail to reject the null hypothesis. Therefore, we do not have sufficient evidence to conclude that stress levels have a significant impact on the number of skipped songs.

```python
stressful_months = ['2023-01', '2023-06', '2024-01', '2024-06']

# Reset index and rename count column
skipped_counts = skipped_counts.reset_index()
skipped_counts.rename(columns={0: 'Skipped Songs'}, inplace=True) 

skipped_counts['Stressful'] = skipped_counts['Month'].astype(str).isin(stressful_months)

stressful_data = skipped_counts[skipped_counts['Stressful'] == True]['Skipped Songs']
non_stressful_data = skipped_counts[skipped_counts['Stressful'] == False]['Skipped Songs']

mean_stressful = stressful_data.mean()
mean_non_stressful = non_stressful_data.mean()

print(f"Mean Skipped Songs (Stressful): {mean_stressful}")
print(f"Mean Skipped Songs (Non-Stressful): {mean_non_stressful}")

t_stat, p_value = ttest_ind(stressful_data, non_stressful_data, equal_var=False)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

categories = ['Stressful', 'Non-Stressful']
means = [mean_stressful, mean_non_stressful]

plt.bar(categories, means, color=['red', 'blue'])
plt.title('Average Number of Skipped Songs')
plt.ylabel('Number of Skipped Songs')
plt.tight_layout()
plt.show()
```

# More info about my listening habbits
I constructed histograms about my top 10 artists and albums.

```python
df['Minutes Played'] = df['ms_played'] / (1000 * 60)  # Convert milliseconds to minutes
artist_listening = df.groupby('master_metadata_album_artist_name')['Minutes Played'].sum()

top_artists = artist_listening.sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
ax = top_artists.plot(kind='bar', color='purple')
plt.title('Top 10 Artists by Listening Time')
plt.ylabel('Total Listening Time (Minutes)')
plt.xlabel('Artist')
plt.xticks(rotation=45)
plt.tight_layout()

# Add exact values above the bars
for index, value in enumerate(top_artists):
    plt.text(index, value + 5, f"{value:.1f}", ha='center', va='bottom', fontsize=10)

plt.show()
```
